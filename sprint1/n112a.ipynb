### Feature Engineering

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

```
df_m = df['ë§¤ì¶œì•¡'].mean()
df['ë§¤ì¶œì•¡']= df['ë§¤ì¶œì•¡'].apply(float)


def get_class(v):
  if df_m *1.1 <= v:
     score = "S"
  elif df_m *1.05 <= v:
     score = "A"
  elif df_m *0.95 <= v:
     score = "B"
  elif df_m *0.9<= v:
     score = "C"
  else:
     score = "D"

  return score


df['Relative Perfomance'] = df['ë§¤ì¶œì•¡'].apply(get_class)
df
```
